```python
# Step 1: Remove text like "(Glassdoor est.)"
df['Salary Estimate Cleaned'] = df['Salary Estimate'].str.replace(r'\(.*?\)', '', regex=True)

# Step 2: Remove symbols
df['Salary Estimate Cleaned'] = df['Salary Estimate Cleaned'].str.replace('$', '', regex=False)
df['Salary Estimate Cleaned'] = df['Salary Estimate Cleaned'].str.replace(' ', '', regex=False)

# Step 3: Split min/max
df[['min_salary_str', 'max_salary_str']] = df['Salary Estimate Cleaned'].str.split('-', expand=True)

# Step 4: Convert strings like "7K" to numeric
def k_to_number(val):
    if pd.isna(val) or val.strip() == '':
        return None
    val = val.upper().replace('K', '')
    try:
        return float(val) * 1000
    except ValueError:
        return None

df['min_salary'] = df['min_salary_str'].apply(k_to_number)
df['max_salary'] = df['max_salary_str'].apply(k_to_number)

# Step 5: Calculate average
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2

# Step 6: Drop helper columns
df.drop(columns=['Salary Estimate', 'min_salary_str', 'max_salary_str'], inplace=True)
```

```python
import pandas as pd

# Load data
df = pd.read_csv("cleaned_analyst_data.csv")

# Standardize text
df['Job Title'] = df['Job Title'].str.strip().str.title()
df['Company Name'] = df['Company Name'].str.strip()

# Extract city/state from Location
df[['City', 'State']] = df['Location'].str.extract(r'^(.*?),\s*([A-Z]{2})$')
```

```python
import numpy as np

# Replace -1 with NaN for relevant columns
cols_with_neg1 = ['Founded', 'Rating', 'Easy Apply', 'Competitors']
df[cols_with_neg1] = df[cols_with_neg1].replace(-1, np.nan)

# Clean Company Name column
df['Company Name'] = df['Company Name'].str.split('\n').str[0]

# Check for inconsistent categories
for col in ['Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue']:
    print(f"Unique values in {col}:", df[col].nunique())

df['Job Description'] = df['Job Description'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Drop redundant column
#df.drop(columns=['Salary Estimate Cleaned'], inplace=True)
```

```python
skills_list = [
    "python", "r", "sql", "excel", "tableau", "power bi",
    "machine learning", "statistics", "data visualization",
    "data analysis", "numpy", "pandas", "matplotlib", "seaborn",
    "scikit-learn", "communication", "problem-solving",
    "spark", "aws", "azure", "big data", "hadoop"
]
```

```python
def extract_skills(Job_Description, skills):
    Job_Description = Job_Description.lower()
    found_skills = [skill for skill in skills if skill in Job_Description]
    return ', '.join(found_skills) if found_skills else 'None'

df['Skills'] = df['Job Description'].apply(lambda x: extract_skills(x, skills_list))
```

```python
# Fill missing 'Rating' with median
df['Rating'] = df['Rating'].fillna(df['Rating'].median())

# Fill 'Founded' with median or a placeholder like 0 or 'Unknown'
df['Founded'] = df['Founded'].fillna(0)  # Or use grouped median by sector/industry

# Drop the 1 row with missing salary info
df = df.dropna(subset=['min_salary', 'avg_salary'])
```

```python
# Save cleaned data
df.to_csv("final_cleaned_data.csv", index=False)
```
